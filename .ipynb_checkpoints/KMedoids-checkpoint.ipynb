```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import accuracy_score, silhouette_score

# Load the dataset
data = pd.read_csv('weather_data.csv')


le = LabelEncoder()
for col in data.columns:
    data[col] = le.fit_transform(data[col])


X = data.drop('description(output)', axis=1)


X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)


kmedoids_model = KMedoids(n_clusters=3, random_state=42) 
kmedoids_model.fit(X_train)


y_pred = kmedoids_model.predict(X_test)


silhouette_avg = silhouette_score(X_test, y_pred)
print(f"Silhouette Score: {silhouette_avg}")




import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


pca = PCA(n_components=2)
X_test_2D = pca.fit_transform(X_test)


plt.figure(figsize=(8, 6))
plt.scatter(X_test_2D[:, 0], X_test_2D[:, 1], c=y_pred, cmap='Set1', s=50, alpha=0.7)
plt.title('K-Medoids Clustering (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()

```
